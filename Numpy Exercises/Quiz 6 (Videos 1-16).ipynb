# Q8: Which two methods do the same thing?

```python
import numpy as np
array1 = np.ones((2,2))
array2 = np.ones((2,2))*2
print("numpy.concatenate(axis=0)", np.concatenate((array1,array2), axis=0))
print("numpy.concatenate(axes=1)", np.concatenate((array1,array2), axis=1))
print("numpy.row_stack", np.row_stack((array1,array2)))
print("numpy.hstack", np.hstack((array1,array2)))
```
